# 變數應用
Q:邊長為3,4,5的三角形，其面積大小為何?  
s=(a+b+c)/2  
sqrt(s(s-a)(s-b)(s-c))

```
import math #数学函数
a,b,c=3,4,5
s=(a+b+c)/2
d=math.sqrt(s*(s-a)*(s-b)*(s-c)) #sqrt平方根
print(d)
```

##list取值


```
a=[1,2,3,4,5,6,7,8,9]
b=a[0:3]#範圍取值，a[0],a[1],a[2]不包含a[3]
print(b)
c=a[0:9:2]#間隔2取值
print(c)
d=a[0:]#a[0]到最後
print(d)
e=a[:9]#從最前面到a[9]
print(e)
f=a[:] #無指定範圍
print(f)
```

```
a=[1,2,3,4,5,6,7,8,9]
del a[7:9]#刪除某範圍，a[7],a[8]
print(a)
print(min(a))#最小值
print(max(a))#最大值
print(a.index(1))#找出1所在的第一個索引值
b=[1,0,0,1,1,1]
print(b.count(1))#1出現的次數
a.reverse()#串列反轉
print(a)
a.sort()#串列由小到大排序，a的值會被改變
print(a)
```
